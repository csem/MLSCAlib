```python
import numpy as np

import os
import matplotlib.pyplot as plt
# Get the current working directory
current_directory = os.getcwd()
print(current_directory)
# Specify the path to your .npy file
results_path = "./Results/"
# file_path = "TestNA_ID_plot-AXIS-na-FAST_GE_SENSI_epochs-2,byte-2,_ASCAD_,Nt_50000,Na_10000,Ns_700.npy"
# file_path = "Test22_HW_1-AXIS-na-FAST_GE_SENSI_epochs-200,byte-2,_ASCAD_,Nt_50000,Na_10000,Ns_700.npy"

# This is good for the MLP_BEST
# file_path = "ASCAD_HW_newPlots-AXIS-na-FAST_GE_SENSI_epochs-200,byte-2,_ASCAD_,Nt_50000,Na_1000,Ns_700.npy"
file_path = "ASCAD_HW_10_r-AXIS-na-FAST_GE_SENSI_epochs-200,byte-2,_ASCAD_,Nt_50000,Na_1000,Ns_700.npy"

# # Read the .npy file
data = np.load(results_path + file_path)

# # Now, 'data' contains the contents of the .npy file
print(data)
print(len(data))
```

```python
# Initialize the result dictionary
result_dict = {}

# Iterate through each record
for record in data:
    name_model = record[3]
    traces = record[0]
    ge = record[1]
    
    # Check if the outer key already exists
    if name_model not in result_dict:
        result_dict[name_model] = {}

    # Check if the inner key already exists and append to the list
    if traces not in result_dict[name_model]:
        result_dict[name_model][traces] = []

    
    # Update the inner dictionary
    result_dict[name_model][traces].append(float(ge))

print(result_dict)
```
